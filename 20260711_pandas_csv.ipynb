```python
def discount(price):
  if price > 250:
    return price * 0.9
  else:
    return price * 0.95
```

```python
df["discounted_price"] = df["price"].apply(discount)
df
```

```python
df.drop(columns="discounted_price")
```

```python
def total_sale(row):
  return row["units"] * row["price"]
```

```python
# Capture timestamps in integer nanoseconds
start_time = time.time_ns()
#Apply function
df['total_sale'] = df[["units", "price"]].apply(total_sale, axis=1)
end_time = time.time_ns()

# Compute the difference
nanoseconds_elapsed_apply = end_time - start_time
print(f"Time difference: {nanoseconds_elapsed_apply} ns")
```

```python
# Capture timestamps in integer nanoseconds
start_time = time.time_ns()
#Vecotorized
df['total_sale'] = df["units"] * df["price"]
end_time = time.time_ns()

# Compute the difference
nanoseconds_elapsed_vectorized = end_time - start_time
print(f"Time difference: {nanoseconds_elapsed_vectorized} ns")
```

```python
print("Vectorized approach is", round(nanoseconds_elapsed_vectorized/nanoseconds_elapsed_apply*100,2), "% faster than Apply",)
```
