```python
import numpy as np
```

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])
print("----------------------")
print ("(A*B) matching slots")
print (A*B)
print("----------------------")
print  ("A dot product B")
print (A.dot(B))
print("----------------------")
print  ("A@B row times column")
print (A @ B)
print("----------------------")
```

```python
A = np.array ([[2,1],[1,3]]) #Capital letter for matrices
b = np.array([50,50]) # Small letter for vectors

prices = np.linalg.solve(A,b)
print ("Samosa price = ", round(prices[0],2))
print ("Chai  price = ", round(prices[1],2))

print (np.allclose((A @ prices),b))
```
