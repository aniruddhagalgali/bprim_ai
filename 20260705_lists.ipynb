```python
import time

def spinning_wheel(duration_seconds=5):
    # The sequence of characters that form the spinning motion
    symbols = ['|', '/', '-', '\\']

    # Calculate how many steps to run based on the delay
    delay = 0.1
    steps = int(duration_seconds / delay)

    print("Charging... ", end="")

    for i in range(steps):
        # Select the correct symbol using modulo
        symbol = symbols[i % len(symbols)]

        # \r resets the cursor to the start of the line, end="" prevents a new line
        print(f"\rCharging... {symbol}", end="", flush=True)

        time.sleep(delay)

    # Clear or complete the loading line
    print("\rCharging... Done!   ")




battery = 20
spinning_wheel(5)
while battery < 100:
  print (battery)
  battery = battery + 20
print("\nBattery fully charged")
```

```python
import getpass

password = "abcd1234"
num_of_tries = 0
while True:
  entered = getpass.getpass("Password")
  num_of_tries = num_of_tries + 1
  if entered == password:
    print("Access granted!")
    break
  if num_of_tries > 2:
    print("3 incorrect attempts. Account locked")
    break
  print("Incorrect password! Attempt ", num_of_tries, " of 3", sep = "")

```

```python
orders = [2,4,8,"bprim","ai",64,128,256.0]

print (orders)

#Slice from start to given element
print (orders[:3])

#Slice from to end
print (orders[3:])

#Slice from to
print (orders[3:4])

#Slice in steps
print (orders[1::3])
```
